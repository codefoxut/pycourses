```python
import pandas_datareader.data as web
dfs = web.DataReader('^SNX', 'stooq')
dfs[:10]
```

```python
df.index[df.Close > df.Open][:10]
df[df.Close > df.Open][:10]
```

```python
dfh = dfs[dfs.Close > dfs.Open][:10]
dfl = dfs[dfs.Close < dfs.Open][:10]
from bokeh.plotting import figure, show, output_file
f = figure(x_axis_type='datetime', width=1000, height=300, title='Candlestick Chart')

hours_12 = 12 * 60 * 60 * 1000
f.rect(dfh.index, 
       (dfh.Open + dfh.Close)/2, 
       hours_12, 
       abs(dfh.Open - dfh.Close),
       fill_color="green", 
       line_color="black")
f.rect(dfl.index, 
       (dfl.Open + dfl.Close)/2, 
       hours_12, 
       abs(dfl.Open - dfl.Close),
       fill_color="red", 
       line_color="black")

output_file("CS.html")
show(f)
```

```python

def inc_dec(c, o):
    if c > o:
        value = "inc"
    elif o > c:
        value = "dec"
    else:
        value = "same"
    return value
dfs["status"] = [inc_dec(c, o) for c, o in zip(dfs.Close, dfs.Open)]
dfs["mid"] = (dfs.Close + dfs.Open) / 2
dfs["height"] = abs(dfs.Close - dfs.Open)
dfs
```

```python
df = dfs[:25]
from bokeh.plotting import figure, show, output_file
f = figure(x_axis_type='datetime', width=1000, height=300, title='Candlestick Chart')

hours_12 = 12 * 60 * 60 * 1000
f.rect(df.index[df.status=="inc"], 
       df.mid[df.status=="inc"], 
       hours_12, 
       df.height[df.status=="inc"],
       fill_color="green", 
       line_color="black")

f.rect(df.index[df.status=="dec"], 
       df.mid[df.status=="dec"], 
       hours_12, 
       df.height[df.status=="dec"],
       fill_color="red", 
       line_color="black")


output_file("CS1.html")
show(f)
```
